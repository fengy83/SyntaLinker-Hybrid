```python
from rdkit import Chem
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./braf_6v2u_m_ks_one.csv')
sdf_1 = Chem.SDMolSupplier('./sp_1uwh_dockingks_gase.sdf')
sdf_2 = Chem.SDMolSupplier('./sp_6v2u_dockingks2_gase.sdf')
w = Chem.SDWriter('./braf_6v2u_m_ks_one.sdf')
mol_ids = list(data['s_m_entry_id'])
file_id = list(data['file_id'])
mol_ids_1 = []
mol_ids_2 = []
for index, mol in enumerate(mol_ids):
    if file_id[index] == 1:
        mol_ids_1.append(str(mol))
    elif file_id[index] == 2:
        mol_ids_2.append(str(mol))
```

```python
for s1 in sdf_1:
    if s1.GetProp("s_m_entry_id") in mol_ids_1:
        w.write(s1)
for s2 in sdf_2:
    if s2.GetProp("s_m_entry_id") in mol_ids_2:
        w.write(s2)
print('done')
```
